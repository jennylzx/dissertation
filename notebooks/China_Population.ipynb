## Plotting

```python
# Plotting
plt.figure(figsize=(10, 5))

plt.plot(df["Year"], df["Total population (ten thousand)"], label='Total Population')
plt.plot(df["Year"], df["Male population (ten thousand)"], label='Male Population')
plt.plot(df["Year"], df["Female population (ten thousand)"], label='Female Population')
plt.plot(df["Year"], df["Urban population (ten thousand)"], label='Urban Population')
plt.plot(df["Year"], df["Rural population (ten thousand)"], label='Rural Population')

plt.xlabel("Year")
plt.ylabel("Population (ten thousand)")
plt.title("Population Over the Years")
plt.legend()
plt.grid(True)

plt.show()
```
