```python
# Convert the 'Number' column to integer type
df['Number'] = df['Number'].astype(int)

# Group the data by CAH Level 1 and Academic Year
grouped_data = df.groupby(['CAH level 1', 'Academic Year']).sum().reset_index()

# Sort the grouped data
sorted_grouped_data = grouped_data.sort_values(['CAH level 1', 'Academic Year'])

# Pivot the data to create separate columns for Male and Female
pivot_data = df.pivot_table(index=['CAH level 1', 'Academic Year'], 
                                                        columns='Sex', 
                                                        values='Number', 
                                                        aggfunc='sum').reset_index()

# Reset column names after pivot
pivot_data.columns.name = ''

# Fill missing values with 0
pivot_data.fillna(0, inplace=True)

# Convert numbers to integers
pivot_data[['Female', 'Male']] = pivot_data[['Female', 'Male']].astype(int)

pivot_data
```
